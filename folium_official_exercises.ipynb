## Exercise 2 — Creating a Map

From the official docs:
> *Here's a basic example of creating a map*

The `location` parameter sets where the map is centred using `[latitude, longitude]`.
Typing `m` on the last line displays the map inline in Jupyter.

```python
# Exercise 2 — Creating a basic map
# Source: https://python-visualization.github.io/folium/latest/getting_started.html

import folium

m = folium.Map(location=(45.5236, -122.6750))

m
```

## Exercise 3 — Choosing a Tileset

From the official docs:

Try changing the `tiles` parameter to see different map backgrounds.

```python
# Exercise 3 — Choosing a tileset
# Source: https://python-visualization.github.io/folium/latest/getting_started.html

import folium

# Try these tileset options:
# "OpenStreetMap"       ← default, detailed roads
# "CartoDB positron"    ← clean light background, good for academic papers
# "CartoDB dark_matter" ← dark background
# "Stamen Terrain"      ← terrain and elevation

m = folium.Map((45.5236, -122.6750), tiles='cartodb positron')

m
```

## Exercise 4 — Adding Markers

From the official docs:
> *There are various marker types, here we start with a simple Marker. You can add a popup and tooltip. You can also pick colors and icons.*

- `tooltip` = text that appears when you **hover** over the marker
- `popup` = text that appears when you **click** the marker
- `icon` = sets the colour and icon style

```python
# Exercise 4 — Adding markers with popups and tooltips
# Source: https://python-visualization.github.io/folium/latest/getting_started.html

import folium

m = folium.Map([45.35, -121.6972], zoom_start=12)

# Marker 1 — cloud icon (default blue colour)
folium.Marker(
    location=[45.3288, -121.6625],
    tooltip='Click me!',
    popup='Mt. Hood Meadows',
    icon=folium.Icon(icon='cloud'),
).add_to(m)

# Marker 2 — green colour
folium.Marker(
    location=[45.3311, -121.7113],
    tooltip='Click me!',
    popup='Timberline Lodge',
    icon=folium.Icon(color='green'),
).add_to(m)

m
```

## Exercise 5 — Vectors: PolyLine

From the official docs:
> *Folium has various vector elements. One example is PolyLine, which can show linear elements on a map. This object can help put emphasis on a trail, a road, or a coastline.*

A `PolyLine` draws a line connecting a series of coordinates.

```python
# Exercise 5 — Drawing a PolyLine (e.g. a trail or route)
# Source: https://python-visualization.github.io/folium/latest/getting_started.html

import folium

m = folium.Map(location=[-71.38, -73.9], zoom_start=11)

trail_coordinates = [
    (-71.351871840295871, -73.655963711222626),
    (-71.374144382613707, -73.719861619751498),
    (-71.391042575973145, -73.784922248007007),
    (-71.400964450973134, -73.851042243124397),
    (-71.402411391077322, -74.050048183880477),
]

folium.PolyLine(trail_coordinates, tooltip='Coast').add_to(m)

m
```

## Exercise 6 — Grouping and Controlling Layers

From the official docs:
> *You can group multiple elements such as markers together in a FeatureGroup. You can select which you want to show by adding a LayerControl to the map.*

- `FeatureGroup` = a named group of markers you can toggle on/off
- `LayerControl` = adds the toggle panel in the top-right corner of the map

```python
# Exercise 6 — FeatureGroup + LayerControl
# Source: https://python-visualization.github.io/folium/latest/getting_started.html

import folium

m = folium.Map((0, 0), zoom_start=7)

# Group 1 — red markers
group_1 = folium.FeatureGroup('first group').add_to(m)
folium.Marker((0, 0), icon=folium.Icon('red')).add_to(group_1)
folium.Marker((1, 0), icon=folium.Icon('red')).add_to(group_1)

# Group 2 — green markers
group_2 = folium.FeatureGroup('second group').add_to(m)
folium.Marker((0, 1), icon=folium.Icon('green')).add_to(group_2)

# Add the layer control toggle panel
folium.LayerControl().add_to(m)

m
```

## Exercise 7 — GeoJSON Overlay

From the official docs:
> *Folium supports both GeoJSON and TopoJSON data in various formats, such as urls, file paths and dictionaries.*

GeoJSON is a standard file format for geographic shapes — like country or province boundaries.

```python
# Exercise 7 — GeoJSON overlay (world country boundaries)
# Source: https://python-visualization.github.io/folium/latest/getting_started.html

import folium
import requests

m = folium.Map(tiles='cartodbpositron')

# Load world country outlines from a URL
geojson_data = requests.get(
    'https://raw.githubusercontent.com/python-visualization/folium-example-data/main/world_countries.json'
).json()

folium.GeoJson(geojson_data, name='hello world').add_to(m)

folium.LayerControl().add_to(m)

m
```

## Exercise 8 — Choropleth Map

From the official docs:
> *Choropleth can be created by binding the data between Pandas DataFrames/Series and Geo/TopoJSON geometries.*

A choropleth colours geographic regions by a data value — darker colour = higher value.
This example shows US unemployment rates by state.

```python
# Exercise 8 — Choropleth map (US unemployment by state)
# Source: https://python-visualization.github.io/folium/latest/getting_started.html

import folium
import pandas
import requests

# Load US state boundaries (GeoJSON)
state_geo = requests.get(
    'https://raw.githubusercontent.com/python-visualization/folium-example-data/main/us_states.json'
).json()

# Load unemployment data (CSV)
state_data = pandas.read_csv(
    'https://raw.githubusercontent.com/python-visualization/folium-example-data/main/us_unemployment_oct_2012.csv'
)

# Build the choropleth map
m = folium.Map(location=[48, -102], zoom_start=3)

folium.Choropleth(
    geo_data=state_geo,
    name='choropleth',
    data=state_data,
    columns=['State', 'Unemployment'],
    key_on='feature.id',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Unemployment Rate (%)',
).add_to(m)

folium.LayerControl().add_to(m)

m
```
